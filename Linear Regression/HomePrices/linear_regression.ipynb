```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
```

```python
df = pd.read_csv(r"homeprices.csv")
```

```python
df.head()
```

```python
plt.xlabel('Area (sq. ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area,df.price, color = "red")
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
area = np.array(df[['area']])
area = area.reshape(-1,1)
area
```

```python
reg.fit(area , df.price)
```

```python

x_pred = np.array(df.area)
x_pred = x_pred.reshape(-1,1)
x_pred

```

```python
y_pred = reg.predict(x_pred)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.xlabel('Area (sq. ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area,df.price, color = "red")
plt.plot(x_pred,y_pred)
plt.show()
```
