```python
def splitKey(key = '2015-05-07 19:52:06.0000003') :

    key = key.split('.')[0]
    dto = datetime.strptime(key, "%Y-%m-%d %H:%M:%S")
    

    # print(f'DATE = {dto}  TIME =')

    # yr
    return dto
    
splitKey('2009-07-17 20:04:56.0000002')
    
```

```python
df

df[['yr','month','day','hr','min','sec']] = df['key'].apply(splitKey)
```
